```python
from splinter import Browser 
from bs4 import BeautifulSoup as soup
import pandas as pd
```

```python
browser = Browser('chrome')
url = 'https://en.wikipedia.org/wiki/Los_Angeles_City_Council'
browser.visit(url)
html = browser.html 
html_soup = soup(html, 'html.parser')
```

```python
#table_headers = html_soup.find_all('th' , class_ = 'headerSort')
table_info = html_soup.find('table', class_ = 'wikitable')
```

```python
df = pd.read_html(str(table_info))
df = pd.DataFrame(df[0])
df['Name'] = df['Name'].replace('Vacant[a]', 'Vacant')
df.drop(columns='Map', inplace = True)
df
```

```python
df.to_csv("Data/CouncilMembers.csv", index=False, header=True)
```
